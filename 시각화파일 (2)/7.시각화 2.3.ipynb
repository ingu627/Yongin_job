### 시각화 2.2 유동인구 정보 시각화(2)

```python
data = pd.read_csv('7.용인시_유동인구.csv')
data
```

```python
data['str_lon']=data['lon'].astype(str)
data['str_lat']=data['lat'].astype(str)
```

```python
data.dtypes
```

```python
data["Full"] = data["str_lon"].map(str) + " " + data["str_lat"]
```

```python
new_data=data.drop(data.columns[0], axis='columns')
```

```python
new_data
```

```python
new_df = pd.DataFrame(new_data.groupby(new_data['Full']).mean())
```

```python
new_df
```

```python
#################여기까지가 측정 지점이 똑같은 데이터를 합친 전처리##################
```

```python
df_tu=new_df.iloc[:,24:26]
df_tu

total_map = pd.DataFrame(
    columns =["total","lon","lat","dong"])

total_map[['lon','lat']]=df_tu
```

```python
total_map
```

```python
df_bn=new_df.iloc[:,7:24]
#df_bn = pd.concat([df_bn,df_tu],axis=1)
total_map['total']=df_bn.mean(axis=1)
total_map
```

```python
df_bn
```

```python
data13=gpd.read_file('13.용인시_행정경계(읍면동).geojson')
data13 = data13.to_crs(epsg='4326')
```

```python
data13
```

```python
#유동인구 80이상인 곳만 히트맵으로 표시
```

```python
total_map2=total_map.iloc[:,0:3]
```

```python
total_map2
```

```python
#유동인구 전처리 - 유동인구가 80이상인 곳만 뽑음
high_total=[]
for i in range(len(total_map2)):
    if total_map2['total'][i] >= 80:
        high_total.append(total_map2.iloc[i])

high_total = pd.DataFrame(high_total)
```

```python
high_total
```

```python
#지도 첫 화면 
center = (37.23, 127.20)
zoom = 11

hm = folium.Map(location = center, 
               zoom_start = zoom,
               tiles = 'http://api.vworld.kr/req/wmts/1.0.0/78B06EC1-6AA1-33E3-8757-90F756C3794A/Base/{z}/{y}/{x}.png',
               attr = '용인시'
              )
##folium.GeoJson(new_data10).add_to(m)
folium.Choropleth(
    geo_data=data13,
    fill_color = 'black',
    fill_opacity =0.3,
    popup=data13.ADM_DR_NM
).add_to(hm)


location =[]
for i in range(len(high_total)):
    location.append([high_total['lat'][i],high_total['lon'][i]])

    
HeatMap(location).add_to(hm)
```
